```python
from bs4 import BeautifulSoup
import requests
```

```python
#declaramos la web de donde extraeremos los datos
web_site = 'https://subslikescript.com/movie/Patriot-4460598'
#con la libreria request enviamos una solicitud a la página 
result = requests.get(web_site)
#nos brinda un resultado y lo almacenamos en result solo el texto
content = result.text
```

```python
#con esto localizamos contenido en una pagina web(código html),declaramos el parse
soup = BeautifulSoup(content, 'lxml')
#visualizamos el contenido
print(soup.prettify())
```

```python
#buscamos un elemento de tipo clase (html) y lo almacenamos en una caja
box = soup.find('article', class_='main-article')
#localizamos el titulo
title = box.find('h1').get_text()
#localizamos los datos
transcrip = box.find('div',class_='full-script').get_text(strip=True, separator='\n') #get_text(strip=True, separator='\n')
```

```python
#visualizamos el titulo
title
```

```python
#visualizamos los datos
transcrip
```

```python
#creamos un archivo .txt
with open(f'{title}.txt', 'w', encoding='utf-8') as file:
    file.write(transcrip)
```
